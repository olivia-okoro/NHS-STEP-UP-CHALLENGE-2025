This report examines four years of NHS antidepressant prescribing data to assess longitudinal trends in prescription volumes and costs across England. The analysis identifies key drivers of prescribing trends, cost inefficiencies, and potential areas for optimization.

## **Key Findings & Insights**


1. Monthly Trends in Antidepressant Prescribing: Volume vs. Cost

   * Total prescription volumes have remained consistently high, with notable
    fluctuations at certain time points.

   * Overall prescribing costs exhibit greater volatility, suggesting price
    changes, policy shifts, or supply chain disruptions impacting total NHS spending.

   * While prescription volume and total cost trends are correlated, they are    not proportional, meaning some drugs disproportionately drive NHS   costs despite lower prescription counts.

2. The Most Prescribed vs. The Most Expensive Antidepressants
   * Sertraline hydrochloride, Fluoxetine, and Citalopram hydrobromide are the most prescribed antidepressants, making up a significant share of total prescriptions.

   * However, Duloxetine and Vortioxetine, while prescribed far less frequently, contribute significantly to total cost, highlighting their higher per-item pricing.
   * Amitriptyline hydrochloride, despite its high prescription volume, has a comparatively low cost impact, suggesting a cost-effective alternative for treatment.

3.  Identifying Cost Inefficiencies

    * Some regions and drugs exhibit high cost per item, despite relatively low prescription volumes, suggesting opportunities for cost optimization.
    * High-cost drugs such as Vortioxetine may require further evaluation for cost-effectiveness relative to clinical outcomes.
    * Generic prescribing patterns should be encouraged, particularly for drugs where cost-effective alternatives exist.


## **Strategic Recommendations**

 1. Analyze cost per item per drug to pinpoint the biggest cost drivers.
 2. Assess seasonal variations in prescribing patterns to optimize NHS procurement strategies.
 3. Compare prescribing trends across NHS regions to identify potential inefficiencies in drug selection or pricing.
 4. Investigate alternative treatment approaches for high-cost antidepressants to ensure sustainable prescribing practices.

By focusing on cost-effective prescribing, price monitoring, and regional prescribing efficiencies, the NHS can optimize spending while maintaining accessibility to essential antidepressant treatments.

## **SUMMARY**

This report presents a comprehensive longitudinal analysis of NHS antidepressant prescribing patterns over a four-year period. The study investigates national and regional trends, prescription volume vs. cost dynamics, cost efficiency of antidepressants, and strategic opportunities for cost optimization.

The findings indicate significant variations in prescribing habits, cost inefficiencies in certain drugs, and opportunities for NHS expenditure optimization without compromising patient care.

## Key Findings & Insights
1.  National & Regional Trends in Antidepressant Prescriptions
Total prescription volumes have remained consistently high, with some seasonal fluctuations.
The Midlands and North East & Yorkshire regions lead in prescription volumes, while London and the North West incur the highest total costs.
Despite a general increase in prescriptions, total cost trends fluctuate significantly, suggesting external factors such as supplier pricing, NHS procurement strategies, or drug policy changes.
2. Prescription Volume vs. Cost Disparities
The most prescribed antidepressants do not always contribute the most to total cost.
Sertraline hydrochloride, Fluoxetine, and Citalopram hydrobromide are the most frequently prescribed drugs, indicating their broad use in treatment plans.
However, high-cost drugs such as Duloxetine and Vortioxetine contribute significantly to total NHS expenditure despite lower prescription counts.
This disproportionate cost impact highlights the importance of monitoring high-cost, low-volume drugs to manage NHS spending effectively.
3. Cost Efficiency & Price Variability Analysis
A few antidepressants contribute disproportionately to NHS costs due to high per-item prices rather than prescription volume.
Drugs such as Tranylcypromine sulfate and Nefazodone hydrochloride have extreme cost per item (£1326.57 and £639.80, respectively), yet their prescribing rates remain low.
The cost per item varies over time, suggesting potential procurement inefficiencies or market-driven price shifts.
4. Cost Efficiency vs. Prescription Volume (Strategic Opportunities)
Drugs with a high cost-to-volume ratio should be re-evaluated for alternative, cost-effective substitutes.
Encouraging the use of generics over high-cost patented antidepressants could significantly reduce NHS expenditures.
Some high-cost drugs may require further assessment for clinical necessity vs. financial feasibility.
Improved forecasting and procurement strategies can help stabilize costs and reduce the impact of price fluctuations.

## **Strategic Recommendations for NHS Cost Optimization**

*  Introduce a cost-per-item monitoring system to detect and manage high-cost drugs in real-time.
*  Encourage cost-effective prescribing by promoting generic equivalents where available.
*  Conduct targeted regional assessments to identify areas with unusually high prescribing costs and evaluate prescribing behaviors.
*  Review high-cost, low-prescription drugs to determine if alternative treatments exist.
*  Implement predictive analytics models to anticipate cost fluctuations and optimize procurement strategies.

##  **Conclusion**
This analysis provides a data-driven foundation for improving NHS prescribing efficiency. While antidepressant prescriptions remain crucial for mental health treatment, there are clear opportunities for cost-saving strategies without compromising patient care.

Through better procurement management, targeted interventions on high-cost drugs, and optimized prescribing behaviors, the NHS can achieve significant financial savings while continuing to provide essential mental health support to patients across England.
